```python
def helper(tau,epsilon,etaa,kappa,delta_tau):
  parameters={}
  X = np.zeros(len(tau))
  P = np.zeros(len(tau))
  PermissionError = np.zeros(len(tau))
  for i in range(len(tau)-1):
    X[i+1] = X[i] + ( (1-X[i])*(1-epsilon*X[i]-P[i]) - (etaa+kappa)*X[i]  )*delta_tau
    P[i+1] = P[i] + epsilon*X[i]*delta_tau
  S = (-epsilon)*X-P+1
  E = -X + 1
  V = (epsilon)*X
  parameters['X'] = X
  parameters['P'] = P
  parameters['S'] = S
  parameters['E'] = E
  parameters['V'] = V
  parameters['tau'] = tau
  return parameters
```

```python
epsilons = np.linspace(0.1,1.5,150)
kappas = np.linspace(0.1,1.5,150)
etaas = np.linspace(0.1,1.5,150)
```

```python
def Best_parameters(tau,epsilons,kappas,etaas,delta_tau,X_true):
  best_epsilon = 0.1
  best_kappa = 0.1
  best_etaa = 0.1
  best_loss = 99999999999
  best_parameters = {}
  for epsilon in epsilons:
    for etaa in etaas:
      for kappa in kappas:
        parameters = helper(tau,epsilon,etaa,kappa,delta_tau)
        loss = np.mean(np.square(parameters['X'] - X_true))
        if loss<best_loss:
          best_loss = loss
          best_parameters = parameters
          best_epsilon = epsilon
          best_etaa = etaa
          best_kappa = kappa
  return best_epsilon,best_etaa,best_kappa,best_parameters
```

```python
best_epsilon,best_etaa,best_kappa,best_parameters = Best_parameters(tau,epsilons,kappas,etaas,0.1,X_true)
```

```python
print(best_epsilon,best_etaa,best_kappa,best_parameters)
```
